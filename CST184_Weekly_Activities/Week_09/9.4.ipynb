```r
#Example 1 
# Load required library
library(ggplot2)

# Define parameters
n <- 10
p <- 0.7
x <- 0:n
y <- dbinom(x, size = n, prob = p)

# Calculate expected value
expected_value <- n * p

# Create data frame for plotting
data <- data.frame(x = x, probability = y)

# Create the plot
ggplot(data, aes(x = x, y = probability)) +
  geom_bar(stat = "identity", fill = "lightgreen", color = "black") +
  geom_vline(xintercept = expected_value, color = "darkgreen", linetype = "dashed", size = 1) +
  labs(title = "Binomial Distribution: Vaccine Recipients",
       subtitle = "Probability of Vaccination in a Group of 10 (p=0.7)",
       x = "Number of People Vaccinated",
       y = "Probability") +
  theme_minimal()
```

```r
#Example 2 
# Load required library
library(ggplot2)

# Define parameters
lambda <- 4
x <- 0:15
y <- dpois(x, lambda = lambda)

# Create data frame for plotting
data <- data.frame(x = x, probability = y)

# Create the plot
ggplot(data, aes(x = x, y = probability)) +
  geom_bar(stat = "identity", fill = "lightblue", color = "black") +
  geom_vline(xintercept = lambda, color = "darkblue", linetype = "dashed", size = 1) +
  labs(title = "Poisson Distribution: Extreme Weather Events",
       subtitle = "Likelihood of Events per Year (λ=4)",
       x = "Number of Extreme Weather Events",
       y = "Probability") +
  theme_minimal()
```

```r
#Example 3
#Load required library
library(ggplot2)

# Define parameters
mean <- 2300
sd <- 400
x <- seq(1000, 3500, length.out = 200)
y <- dnorm(x, mean = mean, sd = sd)

# Create data frame for plotting
data <- data.frame(x = x, density = y)

# Create the plot
ggplot(data, aes(x = x, y = density)) +
  geom_line(color = "red", size = 1) +
  geom_vline(xintercept = mean, color = "darkred", linetype = "dashed", size = 1) +
  labs(title = "Normal Distribution: Daily Calorie Intake",
       subtitle = "Distribution Across Population (Mean=2300 kcal, SD=400)",
       x = "Daily Calorie Intake (kcal)",
       y = "Density") +
  theme_minimal()
```
