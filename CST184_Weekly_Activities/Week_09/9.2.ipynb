```r
install.packages("ggplot2")
library(ggplot2)
```

```r
# Define parameters
mean_temp <- 0     # average anomaly = 0°C
sd_temp <- 1       # standard deviation = 1°C

# Create x-values for the curve
x <- seq(-4, 4, length.out = 200)

# Generate y-values using the density function
y <- dnorm(x, mean = mean_temp, sd = sd_temp)

# Create data frame
df <- data.frame(x, y)

# Plot the normal curve
ggplot(df, aes(x, y)) +
  geom_line(color = "steelblue", linewidth = 1.2) +
  labs(
    title = "Normal Distribution of Temperature Anomalies",
    x = "Temperature Anomaly (°C)",
    y = "Density"
  ) +
  theme_minimal()

# got an error message and changed size to linewidth
```

```r
# Example: higher mean and wider spread
mean_income <- 30000
sd_income <- 10000

x <- seq(0, 120000, length.out = 10000)
y <- dnorm(x, mean = mean_income, sd = sd_income)
df <- data.frame(x, y)

ggplot(df, aes(x, y)) +
  geom_line(color = "firebrick", size = 1.2) +
  labs(
    title = "Modified Normal Distribution (Mean = 30000, SD = 10000)",
    x = "Income",
    y = "Density"
  ) +
  theme_minimal()

#when i expanded it by changing the x axis to a higher number i saw it is also represented as a bellcurve
```
