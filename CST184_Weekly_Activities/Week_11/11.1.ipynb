```r
# Neutral prior: Beta(2, 2)
alpha_prior <- 2
beta_prior <- 2
```

```r
# Observed data
successes <- 38
failures <- 12
```

```r
# Posterior parameters
alpha_post <- alpha_prior + successes
beta_post <- beta_prior + failures
```

```r
posterior <- rbeta(10000, alpha_post, beta_post)
```

```r
hist(posterior, breaks = 50, col = "skyblue",
     main = "Posterior Distribution",
     xlab = "Proportion of Households with Access")
```

```r
mean(posterior > 0.7)
```
