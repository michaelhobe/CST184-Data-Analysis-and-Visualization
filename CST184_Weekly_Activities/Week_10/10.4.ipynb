```r
library(ggplot2)


ggplot(health_df, aes(x = sugar_intake, y = obesity_rate)) +
  geom_point(color = "darkgreen") +
  geom_smooth(method = "lm", se = TRUE, color = "red") +
  labs(
    title = "Regression: Sugar Intake vs Obesity Rate",
    x = "Daily Sugar Intake (g)",
    y = "Obesity Rate (%)"
  ) +
  theme_minimal()
```
