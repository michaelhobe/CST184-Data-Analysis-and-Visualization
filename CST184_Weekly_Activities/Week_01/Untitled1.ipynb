```python
import pandas as pd

df = pd.read_csv("renewable_energy.csv", encoding="utf-8")
df.head()
```

```python
high_renewable = df[df["%Renewable"] > 50]
high_renewable
```

```python
high_set = set(high_renewable["Country"])
print("Countries with >50% renewable energy:", high_set)
print("Cardinality:", len(high_set))
```

```python
emissions_dict = dict(zip(df["Country"], df["CO2_Emissions"]))
print(emissions_dict["Brazil"])
```

```python
import matplotlib.pyplot as plt

plt.bar(emissions_dict.keys(), emissions_dict.values())
plt.xticks(rotation=45)
plt.title("COâ‚‚ Emissions per Capita by Country")
plt.ylabel("Tonnes")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df["Population"], df["%Renewable"])

for i in range(len(df)):
    plt.text(df["Population"][i], df["%Renewable"][i], df["Country"][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
asia_df = df[df["Region"] == "Asia"]
```

```python
plt.scatter(asia_df["Population"], asia_df["%Renewable"], color="green")
for i in range(len(asia_df)):
    plt.text(asia_df["Population"].iloc[i], asia_df["%Renewable"].iloc[i], asia_df["Country"].iloc[i], fontsize=8)
plt.title("Population vs % Renewable Energy Use (Asia)")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```
