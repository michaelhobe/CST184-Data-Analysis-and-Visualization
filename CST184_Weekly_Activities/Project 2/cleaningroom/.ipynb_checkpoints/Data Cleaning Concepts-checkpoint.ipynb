```python
# 1. Inspect the Dataset Structure
# Purpose: Understand the dataset’s size, column types, and initial missing values.

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load CSVs
df = pd.read_csv("owid_climate_data.csv")

# Display basic info
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
print("\nDataset shape:", df.shape)
```

```python
# 2. Check for Missing Values
# Purpose: Identify and quantify missing data in each column.

# Count missing values per column
print("Missing values per column:")
print(df.isnull().sum())
# Percentage of missing values
print("\nPercentage missing:")
print(df.isnull().mean() * 100)
```

```python
# 3. Visualize Missing Values
# Purpose: Graphically explore missing data patterns.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# 4. Check Unique Values in Categorical Columns
# Purpose: Validate country names for consistency and errors.

# List unique countries
print("Unique countries:", df['country'].nunique())
print("\nCountry names:")
print(sorted(df['country'].unique()))
# Check for similar names (e.g., typos)
from fuzzywuzzy import process
countries = df['country'].unique()
for country in countries:
    matches = process.extract(country, countries, limit=3)
    print(f"Similar to {country}: {matches}")
```

```python
# 5. Check for Unexpected Categories
# Purpose: Identify non-country entries in the country column (e.g., regions or aggregates).

# Flag entries that might not be countries
non_countries = ['World', 'Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America']
print("Potential non-country entries:")
print(df[df['country'].isin(non_countries)][['country', 'year']].drop_duplicates())
```

```python
# 6. Validate year Values
# Purpose: Ensure year contains reasonable values.

# Check range of years
print("Year range:", df['year'].min(), "to", df['year'].max())
# Check for invalid years (e.g., negative or future years)
invalid_years = df[~df['year'].between(1800, 2025)]['year'].unique()
print("Invalid years:", invalid_years)
```

```python
# 7. Check for Zero Values in co2_emissions and temperature
# Purpose: Identify zero values, which may indicate missing data or errors.

# Count zero values
print("Zero CO2 emissions:", (df['co2'] == 0).sum())
print("Zero temperature:", (df['temperature'] == 0).sum())
# Inspect rows with zero CO2
print("\nRows with zero CO2:")
print(df[df['co2'] == 0][['country', 'year', 'co2']])
```

```python
# 8. Check for Negative Values
# Purpose: Ensure co2_emissions and temperature don’t have invalid negative values.

# Check for negative values
print("Negative CO2 emissions:", (df['co2'] < 0).sum())
print("Negative temperatures:", (df['temperature'] < 0).sum())
# Inspect negative CO2 rows
print("\nRows with negative CO2:")
print(df[df['co2'] < 0][['country', 'year', 'co2']])
```

```python
# 9. Detect Outliers in Numerical Columns
# Purpose: Identify extreme values in co2_emissions and temperature using the IQR method.

# Function to detect outliers
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Apply to CO2 and temperature
co2_outliers = find_outliers(df, 'co2')
temp_outliers = find_outliers(df, 'temperature')
print("CO2 outliers:", co2_outliers[['country', 'year', 'co2']])
print("Temperature outliers:", temp_outliers[['country', 'year', 'temperature']])
```

```python
# 10. Visualize Numerical Distributions
# Purpose: Plot histograms to spot anomalies in co2_emissions and temperature.

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['co2'], bins=30, kde=True)
plt.title('CO2 Emissions Distribution')
plt.subplot(1, 2, 2)
sns.histplot(df['temperature'], bins=30, kde=True)
plt.title('Temperature Distribution')
plt.tight_layout()
plt.show()
```

```python
# 11. Check for Duplicates
# Purpose: Identify duplicate rows or country-year combinations.

# Check for exact duplicates
print("Duplicate rows:", df.duplicated().sum())
# Check for duplicate country-year pairs
print("Duplicate country-year pairs:", df.duplicated(subset=['country', 'year']).sum())
# Inspect duplicates
print("\nDuplicate country-year pairs:")
print(df[df.duplicated(subset=['country', 'year'], keep=False)][['country', 'year']])
```

```python
# 12. Validate Data Types
# Purpose: Ensure columns have correct data types.

# Check current data types
print("Data types:\n", df.dtypes)

# Convert if needed (example) - DO NOT USE. NOT REQUIRED SINCE ALL DATA IS CORRECT TYPE FOR EACH COLUMN.
# df['year'] = df['year'].astype(int)
# df['country'] = df['country'].astype(str)
# df['co2'] = df['co2'].astype(float)
# df['temperature'] = df['temperature'].astype(float)
# print("\nUpdated data types:\n", df.dtypes)
```

```python
# 13. Check Temporal Consistency
# Purpose: Ensure each country has consistent year coverage.

# Count years per country
year_counts = df.groupby('country')['year'].nunique()
print("Years per country:\n", year_counts)
# Flag countries with sparse data
sparse_countries = year_counts[year_counts < year_counts.max() * 0.5]
print("\nCountries with sparse years:\n", sparse_countries)
```

```python
# 14. Validate Temperature Ranges
# Purpose: Ensure temperature values are climatologically plausible.

# Check temperature range
print("Temperature range:", df['temperature'].min(), "to", df['temperature'].max())
# Flag extreme temperatures
extreme_temps = df[(df['temperature'] < -30) | (df['temperature'] > 40)]
print("Extreme temperatures:\n", extreme_temps[['country', 'year', 'temperature']])
```

```python
# 15. Check CO2 Emissions Reasonableness
# Purpose: Ensure co2_emissions values are plausible.

# Check range
print("CO2 emissions range:", df['co2'].min(), "to", df['co2'].max())
# Flag extreme CO2 values
extreme_co2 = df[df['co2'] > 50]  # Arbitrary threshold; adjust based on data
print("Extreme CO2 emissions:\n", extreme_co2[['country', 'year', 'co2']])
```

```python
# 16. Check for Unit Consistency
# Purpose: Verify if temperature and co2_emissions units are consistent across countries.
    
# Group by country and check standard deviation of temperature
temp_std = df.groupby('country')['temperature'].std()
print("Temperature std dev by country:\n", temp_std.sort_values(ascending=False))
# Flag countries with high variability
print("\nHigh temperature variability:\n", temp_std[temp_std > 5])
```
